##  a) Function: KO variable corr vector

```python
def KOvar_corr_vector(Env_chem, out_var):    #Gene_abundT,Env_chem)

    ### 1) Make [Genes, Env] Correlation matrix:  merge, corr, subset NR
    Gene_abundT = Gene_abund.T                                          # Transpose before join
    Genes_env = Env_chem.join(Gene_abundT, how='right')                 # JOIN genes and metadata
    Genes_env_corr = Genes_env.corr()                                   # Make CORR matrix

    # Subset Corr matrix by [Genes, Env]                           
    n_enVars = len(env_chem_List)                                       # get n Env_vars
    Genes_env_CORR = Genes_env_corr.iloc[n_enVars:,:n_enVars]           # reduce Corr.M to include only [Genes, Env] 

    # Prep corr matrix for merge w. ontol, set KO col, drop index
    Genes_env_CORR['KO']= Genes_env_CORR.index                          # M1: Set KO as column, drop index
    Genes_env_CORr = Genes_env_CORR.reset_index(drop=True)              # Drop index      ## Genes_env_CORr.head() # Genes_env_CORr.shape

    ## Select var of interest
    #var = 'pH'
    corr_vector = Genes_env_CORr[[out_var,'KO']]
    return corr_vector
```

## b) Function: subset KO (lists) using correlation Cutoffs ( r >)
Work done not to call exec from within filter

```python
def KOvar_corr_cutoff(corVect, r_cut):
    # Cuts off a correlation vector at user defined r
    out_var = corVect.columns[0] 
    corVect['abs_var'] = abs(corVect[out_var])                           # Get absolute value    # append Abs. Value to matrix      # Genes_env_ontol_Corr.head()  # Genes_env_ontol_Corr.shape

    # Cutoff at data |r| > cut                                                              
    corV_r_cut= corVect['abs_var'] > r_cut                               # Cut at abs True?      # Needed for function call
    corV_r_cut2 = corVect[corV_r_cut]                                    # True subset 
    corV_r_cutKO = pd.DataFrame(corV_r_cut2['KO'])                       # only KOs
    return(corV_r_cutKO)

    #KO_var = corVect.columns[1]                                         # Get col of interest
    #abs_var = "abs_"+ out_var                                           # abs_var
    #corVect[abs_var] = abs(corVect[out_var])                            # append Abs. Value to matrix      # Genes_env_ontol_Corr.head()  # Genes_env_ontol_Corr.shape
```

### c) get TOP 10 gene (+, -) list for plotting

#### get top10 function:

```python
#def get_top_10_lists(corrXvar)  # should update to reflect input data--simplify naming
#corrXvar= 'Oligo_CH4'
#CorrsOntol = CH4_Sal_all_corr_O

def get_top10_corrs(corrXvar, CorrsOntol):  #corrX var is str. column, CorrOntolS is DF of KO corrs + ontology
    CorrsOntolS = CorrsOntol.sort_values(corrXvar, ascending=False)     # SORT by corrXvar corr val
    Top10_ont = CorrsOntolS.head(10)                                            # get top 10 (+) correlated
    Bottom10 = CorrsOntolS.tail(10)                                             # get bottom 10 (-) correlated
    Bottom10_r = Bottom10.sort_values(corrXvar, ascending=True)                         # Reverse bottom 10 order                # Bottom10_r
    TopBot10 = Top10_ont.append(Bottom10_r)                                             # Combine top and bottom 10 vars 
    TopBot10smN = pd.DataFrame(TopBot10[['sm_name', corrXvar]])                         # Make DF of SM and Corr   # TopBot10smN
    #print(TopBot10smN)
    return(TopBot10smN)
```

### run top10 function for Corr based on subsets

```python
# get Oligo_CH4 top 10 KOs
#CH4_Sal_all_corr_O.columns()
All_CH4_top10 = get_top10_corrs('All_CH4', CH4_Sal_all_corr_O)
Oligo_CH4_top10 = get_top10_corrs('Oligo_CH4', CH4_Sal_all_corr_O)

All_Sal_top10 = get_top10_corrs('All_Sal', CH4_Sal_all_corr_O)
Ref_Sal_top10 =  get_top10_corrs('Ref_Sal', CH4_Sal_all_corr_O)    # Ref_Sal_top10 
```

## e) Elaborate function to take sample subsets

```python
# Adapt plotting function for Rest, Ref series from Plotting DEV:
# Plots two scatters (Rest, Ref) on same axis, using color vectors per point
def Rest_ref_scatter(y,x,Ref_dat, Rest_dat, axis):
    sns.regplot(y=y, x = x, fit_reg=False, marker ='o', data=Ref_dat, ax=axis, 
      scatter_kws={'facecolors':Ref_dat['color'],'alpha':0.8})  
    sns.regplot(y=y, x = x, fit_reg=False, marker ='^', data=Rest_dat, ax=axis, 
      scatter_kws={'facecolors':Rest_dat['color'],'alpha':0.8})
    #sns.despine()    
```

```python
# Function for sorting TOP 10 lists by Ontology order for interpretation
def Top10_Ontol_sort(Top10):
    Top10_O = pd.merge(Top10, gene_ontol_colors, on='sm_name')
    Top10_OS = Top10_O.sort_values('Ont_order')
    Top10_OS.index = Top10_OS['sm_name']
    return(Top10_OS)
```

```python
# Make 2x5 Top10 GRID FUNCTION that PRINTS output
def Top_10_scatter_subsets(y, x, Ref_dat, Rest_dat, filename):
    from matplotlib.backends.backend_pdf import PdfPages
    #pp = PdfPages(filename)                # Initialize the pdf file
    
    fig = plt.figure(figsize=(12, 6))        # Define plot
    gs = gridspec.GridSpec(2, 5)                # Declare grid dimension   
    gs.update(left=0, right=1, wspace=0.45)     # Fine tune panel spacing

    ax1 = plt.subplot(gs[0, 0])                 # 1st axis
    ax2 = plt.subplot(gs[0, 1])                 # 2nd axis
    ax3 = plt.subplot(gs[0, 2])                 # 3rd axis
    ax4 = plt.subplot(gs[0, 3])                 # 4th axis
    ax5 = plt.subplot(gs[0, 4])                 # 5th axis

    ax6 = plt.subplot(gs[1, 0])                 
    ax7 = plt.subplot(gs[1, 1])                 
    ax8 = plt.subplot(gs[1, 2])                 
    ax9 = plt.subplot(gs[1, 3])                 
    ax10 = plt.subplot(gs[1, 4])

    Rest_ref_scatter(y,x.sm_name[0],Ref_dat, Rest_dat, ax1)
    Rest_ref_scatter(y,x.sm_name[1],Ref_dat, Rest_dat, ax2)
    Rest_ref_scatter(y,x.sm_name[2],Ref_dat, Rest_dat, ax3)
    Rest_ref_scatter(y,x.sm_name[3],Ref_dat, Rest_dat, ax4)
    Rest_ref_scatter(y,x.sm_name[4],Ref_dat, Rest_dat, ax5)

    Rest_ref_scatter(y,x.sm_name[5],Ref_dat, Rest_dat, ax6)
    Rest_ref_scatter(y,x.sm_name[6],Ref_dat, Rest_dat, ax7)
    Rest_ref_scatter(y,x.sm_name[7],Ref_dat, Rest_dat, ax8)
    Rest_ref_scatter(y,x.sm_name[8],Ref_dat, Rest_dat, ax9)
    Rest_ref_scatter(y,x.sm_name[9],Ref_dat, Rest_dat, ax10)
    #plt.show()

    gs.tight_layout(fig)                    # Make tight layout
    #pp.savefig()                            # Save the figure PDF created at top, else add file name .pdf  or #plt.show() #('soil_CH4_Sal_corr_test.pdf')
    #pp.close()   # Close the file
```
