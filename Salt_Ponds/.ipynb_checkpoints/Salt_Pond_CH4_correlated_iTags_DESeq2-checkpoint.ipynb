### Clean Taxonomy strings / cols, contaminants, unknowns - Make MODULE

```r
clean_OTU_taxon_to_Physeq = function(OTUin) {
    # Function takes an OTU table with Consensus.lineage, returns OTU table with addtl 6 taxonomic levels for Phyloseq
    # fxn also drops poorly ID'd taxa, cleans some misc delimiters
    # importantly, fxn fills in missing (unknown) taxonomy at each level with next highest taxon, e.g. Nitrospira_CL
    
    ######################################################
    ## 1) Split strings, clean delimiters, misc text ## 

    # Split Taxonomy strings
    TaxSplit <- str_split_fixed(as.character(OTUin$Consensus.lineage),';', 6)
    colnames(TaxSplit)<-c("Kingdom", "Phylum", "Class", "Order", "Family", "Genus")       # previously made phyloseq b4 cleaning # TaxTable<-tax_table(NameNumber)   # Make Phyloseq table # head(TaxTable) # attributes(TaxSplit)

    # Clean string delimiters
    TaxSplit <- data.frame(TaxSplit)
    TaxSplit<-as.data.frame(sapply(TaxSplit, gsub,pattern="k__",replace=""))
    TaxSplit<-as.data.frame(sapply(TaxSplit, gsub,pattern="p__",replace=""))
    TaxSplit<-as.data.frame(sapply(TaxSplit, gsub,pattern="c__",replace=""))
    TaxSplit<-as.data.frame(sapply(TaxSplit, gsub,pattern="o__",replace=""))
    TaxSplit<-as.data.frame(sapply(TaxSplit, gsub,pattern="f__",replace=""))
    TaxSplit<-as.data.frame(sapply(TaxSplit, gsub,pattern="g__",replace=""))             # TaxSplit

    # Clean misc. text: 
    TaxSplit<-as.data.frame(sapply(TaxSplit, gsub,pattern="\\s\\(class\\)",replace=""))  # Drop " (class)", note double esc.
    TaxSplit<-as.data.frame(sapply(TaxSplit, gsub,pattern="Candidatus",replace="Cand.")) # "Canditatus" genera -> "Cand."
    row.names(TaxSplit) <-row.names(OTUin)                                               # Add back original OTU numbers                  # head(TaxSplit) #    # TaxSplit

    ######################################################
    ## 2) Drop unwanted groups, poorly ID seq's ## 

    # Drop chloroplast & mitochondrial reads
    TaxSplit2 <-data.frame(TaxSplit)                                                     # not needed anymore, but leave for numbering # levels(TaxSplit2$Kingdom)  # TaxSplit2$Genus
    TaxSplit3 <- TaxSplit2[!grepl("chloroplast", TaxSplit2$Kingdom),]                    # drop chloropl.   
    TaxSplit4 <- TaxSplit3[!grepl("mitochondria", TaxSplit3$Kingdom),]                   # drop mitochon.             # dim(TaxSplit2) # dim(TaxSplit4)

    # Drop reads with no phylum
    TaxSplit5 <- TaxSplit4[!grepl("BacteriaKI", TaxSplit4$Phylum),]                      # assigned KI only 
    TaxSplit5 <- TaxSplit4[!grepl("ArchaeaKI", TaxSplit4$Phylum),]                       # assigned KI only 
    levels(TaxSplit5$Phylum)[levels(TaxSplit5$Phylum)==""] <- NA_character_              # assign NA to Empty Phyla #  Magic, see is.na fxn doc
    TaxClean0 <- TaxSplit5[!is.na(TaxSplit5$Phylum),]                                    # drop NA rows             #  dim(TaxClean0); dim(TaxSplit2)#levels(TaxClean$Phylum)  #  # TaxClean0

    ######################################################
    ## 3) Fill missing taxonomic levels, rebuild Tax Table ## 

    # Define Tax rank cols (needed below, avoiding attach method)
    OTU <- row.names(TaxClean0)
    Kingdom <- TaxClean0$Kingdom
    Phylum <- TaxClean0$Phylum
    class <- TaxClean0$Class
    order <- TaxClean0$Order
    family <- TaxClean0$Family
    genus <- TaxClean0$Genus

    # Fill missing, note capitalization for new vars, fill is sequential using new if missing        # Phylum<-ifelse(TaxClean0$phylum=="",paste(Kingdom,"KI",sep=""),paste(phylum))      # already deleted these
    Class<-ifelse(TaxClean0$Class=="",paste(Phylum,"PH",sep=""),paste(class))
    Order<-ifelse(TaxClean0$Order=="",paste(Class,"CL",sep=""),paste(order))
    Family<-ifelse(TaxClean0$Family=="",paste(Order,"OR",sep=""),paste(family))
    Genus<-ifelse(TaxClean0$Genus=="",paste(Family,"FA",sep=""),paste(genus))

    # Rejoin Taxonomy DS elements
    TaxRedo <-data.frame(OTU, Kingdom,Phylum,Class,Order,Family,Genus)                               # length(OTU);length(Kingdom);length(Phylum);length(Class);length(Order);length(Family);length(Genus)

    # Cleanup multiple sequential tags
    TaxRedo <-as.data.frame(sapply(TaxRedo, gsub,pattern="PHCLORFA",replace="PH"))
    TaxRedo <-as.data.frame(sapply(TaxRedo, gsub,pattern="PHCLOR",replace="PH"))
    TaxRedo <-as.data.frame(sapply(TaxRedo, gsub,pattern="PHCL",replace="PH"))
    TaxRedo <-as.data.frame(sapply(TaxRedo, gsub,pattern="CLORFA",replace="CL"))
    TaxRedo <-as.data.frame(sapply(TaxRedo, gsub,pattern="CLOR",replace="CL"))
    TaxRedo <-as.data.frame(sapply(TaxRedo, gsub,pattern="ORFA",replace="OR"))

    # add OTU as row names
    row.names(TaxRedo)<-TaxRedo$OTU                                                                  # TaxRedo$Genus # dim(TaxRedo)  # TaxRedo

    ######################################################
    ## 4) Prune OTU table by cleaned taxonomy ##
    OTUin$OTU <-row.names(OTUin)                                          # Get origninal OTU table (OTUin), add OTU column       #head(OTUin)
    KeepOTU <- data.frame(OTU=TaxRedo$OTU)                                # Gather OTU numbers from cleaned taxonomy
    row.names(KeepOTU)<-KeepOTU$OTU                                       # Add OTU row names for later.                # KeepOTU

    cleanOTU <- merge(OTUin,TaxRedo, by="OTU")                            # Merge KeepOTUs with original OTU table      # keeping taxonomy for check...  # cleanOTU
    row.names(cleanOTU) <- cleanOTU$OTU                                   # Make OTUs row.names after merge
    return(cleanOTU)
}
```

```r
# Run clean OTU function on OTUin data
cleanOTU = clean_OTU_taxon_to_Physeq(OTUin)
#dim(Test); head(Test) 
```

## Make function for DESeq VST CPM from Physeq 

```r
# Note physeq_2_deseq2 forces use of control treatment, though this effect is overwritten by VST calcs.
# VST calcs not possible without this step by this method

# DESeq2_VST_CPM = function(treatment, level){}
### Make PHYLOSEQ object                                      # suppressMessages() -- not working
#physeq = phyloseq(OTU,Map,TaxTable)                          # physeq 

# Assign control treatment, phyloseq patch before DESeq
sample_data(physeq)$Treat <- relevel(as.factor(get_variable(physeq, "Treat")), ref="Historic")
OTU_phy2des <- phyloseq_to_deseq2(physeq, ~ Treat)                     # phyloseq out TO DESeq2 
OTU_phy2des <- DESeq(OTU_phy2des, test="Wald", fitType="parametric")   # Wald test                                        #OTU_phy2des 

# Estimate size factors and Dispersion, get VST data
OTU_phy2des_vs <- estimateSizeFactors(OTU_phy2des)           # Est. size factors
OTU_phy2des_vs <- estimateDispersions(OTU_phy2des)           # Est. dispersions 
OTU_phy2des_VST <- getVarianceStabilizedData(OTU_phy2des)    # Calc variance stabilized data                   # OTU_phy2des_VST

## REPLACE VST data with COUNTS
OTU_vs_counts <- counts(OTU_phy2des_vs, normalized=TRUE)     # get counts                                      # head(KO_vs_counts) 
OTU_table_VST <- data.frame(OTU_vs_counts)                   # make df, shorten name

# Scale OTU VST table counts to CPM (Counts per million)
sampTots<-colSums(OTU_table_VST)                             # Get sample totals                               # [,1:ncol(OTU_table_VST)-1]) 
OTU_VST_CPM<-sweep(OTU_table_VST*1000000,2,sampTots,'/')     # Sweep matrix by:[div. by samp total * 1e+06]    # colSums(OTU_VST_CPM)#head(OTU_VST_CPM)  # OTU_VST_CPM
OTU_VST_CPM$OTU <- row.names(OTU_VST_CPM)                    # Add OTU as column for later merges              # # head(OTU_VST_CPM) #
# colSums(OTU_VST_CPM[,1:(ncol(OTU_VST_CPM)-1)],)            # check sums                                      # colSums(OTU_VST_CPM) # KO_VST_CPM  

print('Deseq2 VST CPM finished computing')
```

# Correlations module

### Get environment variable

```r
# Get Environmental variable of interest, output as ENV                                                    # colnames(map_iTag)
CH4 <- map_iTag$CH4_umol_m2_d                                         # Short name
CH4_logn1 <- CH4 # log10(CH4)                                                 # SKIP log 10, contains NA due to negative CH4
CH4_logn1 <- data.frame(CH4_logn1=log10(CH4 - (1.05*min(CH4))))       # Get log n+1 data. since negative, add 5% extra so non-zero
CH4_logn1$Sample <-map_iTag$Sample                                    # add sample col for merge
CH4_logn1$Indx <-map_iTag$New_index                                   # add New_index for sort post merge  # CH4_logn1  # CH4 

# ABSTRACT this variable
env <- CH4_logn1
```

```r
#  MAKE FUNCTION -- 
### Get DESeq normalized log2 count data Matrix                          # From OTU_VST_CPM
# New OTU ids needed for transpose                                     
row.names(OTU_VST_CPM)<-paste0("otu_",OTU_VST_CPM$OTU)                   # added to matrix (for colnames)
row.names(OTU_VST_CPM_tax)<-paste0("otu_",OTU_VST_CPM_tax$OTU)           # added to Taxonomy mtx (for taxon recovery)        #OTU_VST_CPM  # dim(OTU_VST_CPM_tax)                               
                 
# Transpose and Log2 transform OTU table
OTU_VST_CPMd <-OTU_VST_CPM[,2:(ncol(OTU_VST_CPM)-1)]                     # Drop OTU,              # OTU_VST_CPM
OTU_VST_CPM_t <-data.frame(t(OTU_VST_CPMd))                              # transpose
OTU_CPM_log2_t <- log2(OTU_VST_CPM_t)                                    # LOG 2 transform data
# OTU_VST_CPM_t
OTU_CPM_log2_t$Sample <- row.names(OTU_CPM_log2_t)                       # add samples col        # OTU_VST_CPM_t

# Merge by sample, order by site
envV_OTU_CPM <- merge(env, OTU_CPM_log2_t, by="Sample")                  # merge
row.names(envV_OTU_CPM) <- envV_OTU_CPM$Sample                           # Sample as row.name 
envV_OTU_CPM <- envV_OTU_CPM[order(envV_OTU_CPM$Indx),]                  # order by Index
envV_OTU_CPM <- envV_OTU_CPM[,-which(names(envV_OTU_CPM) == "Sample")]   # drop Sample var
envV_OTU_CPM <- envV_OTU_CPM[,-which(names(envV_OTU_CPM) == "Indx")]     # drop Sample var         # CH4_OTU_CPM
envV_OTU_CPMm <- data.matrix(envV_OTU_CPM)                               # Get Env-OTU Corr matrix

#row.names(envV_OTU_CPMm)
# envV_OTU_CPMm <- envV_OTU_CPMm[12:23,]       # only salt pond SITES
# envV_OTU_CPMm <- envV_OTU_CPMm[6:23,]       # no Ref SITES
# row.names(envV_OTU_CPMm)
```

```r
# Get correlation square matrix for r and P values, only use first row
envOTUcorr.M <- rcorr(envV_OTU_CPMm, type=c("spearman"))                          # Spearman correlations, non-param
envOTUcorr.m <- data.frame("r"=envOTUcorr.M$r[-1,1], "P"=envOTUcorr.M$P[-1,1])    # combine r and P
envOTUcorr.m$P.fdr <- p.adjust(envOTUcorr.m$P, method="BH")                       # correct for false discov. rate
#envOTUcorr.m
```

```r
# Set correlation cutoffs for P-value and r.
P_cut = 0.05
r_cut = 0.75

# Filter to make list of genes correlated to C turnover 
envOTUcorr.Pfilt <- subset(envOTUcorr.m, P.fdr < P_cut)                          # Cutoff on P values
envOTUcorr.P.r.filt <- subset(envOTUcorr.Pfilt, abs(r) > r_cut)               # cutoff in r

# Separate plus and minus
envOTUcorr_plus <- subset(envOTUcorr.P.r.filt, r > 0) 
envOTUcorr_minus <- subset(envOTUcorr.P.r.filt, r < 0) 
envOTUcorr_both <- rbind(envOTUcorr_plus, envOTUcorr_minus) 

# Get lists 
envOTUcorr_plus.list <- data.frame("otu_name"= row.names(envOTUcorr_plus))        # extract df list of corr. genes     # envOTUcorr.P.r.filt
envOTUcorr_minus.list <- data.frame("otu_name"= row.names(envOTUcorr_minus))      # extract df list of corr. genes     # envOTUcorr.P.r.filt
envOTUcorr.list <- unique(rbind(envOTUcorr_plus.list, envOTUcorr_minus.list))     # combine lists                     #data.frame("otu_name"= row.names(envOTUcorr.P.r.filt))             # extract df list of corr. genes     # envOTUcorr.P.r.filt
```

```r
# FUNCTION to Get back original OTU numbers from "otu_n" colnames 
OTU_Name_remap <- function (envOTUcorr.list) {
    
    # Get name mapping 
    otu_name <- row.names(OTU_VST_CPM)                      # get OTU name
    OTU_strip <- data.frame(otu_name, OTU=OTU_VST_CPM$OTU)  # Make data frame   # OTU_strip

    # Merge corr OTUs
    CorrOTU_mrg <- merge(envOTUcorr.list, OTU_strip, all.x=TRUE, by="otu_name") # dim(CorrOTU_mrg)  # dim(envOTUcorr.list)
    CorrOTU <-data.frame(OTU=CorrOTU_mrg$OTU)
    
return(CorrOTU)    
}
```

```r
# Get plus and minus lists as char, number only
envOTUcorr_plusL <-OTU_Name_remap(envOTUcorr_plus.list)            # positive corr
envOTUcorr_minusL <-OTU_Name_remap(envOTUcorr_minus.list)          # negative corr
envOTUcorr_bothL <- unique(rbind(envOTUcorr_plusL, envOTUcorr_minusL))    # combine both lists
# envOTUcorr_bothL
```

```r
## Get DFs with full taxonomy        # envOTUcorr_both
OTU_Tax_only <- subset(OTU_VST_CPM_tax, select = c("Consensus.lineage","Kingdom","Phylum","Class","Order","Family","Genus"))
OTU_Tax_only$otu_name <- row.names(OTU_Tax_only)                       # add otu_name col from rowname                #OTU_Tax_only

# Get ENV corr both and merge with Taxonomy
envOTUcorr_both$otu_name <- row.names(envOTUcorr_both)                 # add otu_name col from rowname 
OTU_Tax_envCorr <- merge(envOTUcorr_both, OTU_Tax_only, by="otu_name") # merge $OTU<- row.names(envOTUcorr)
OTU_Tax_envCorr <- OTU_Tax_envCorr[order(OTU_Tax_envCorr$r),]          # sort by r
#OTU_Tax_envCorr

# Subset that are archaea
Arch_corr <-OTU_Tax_envCorr[which(OTU_Tax_envCorr$Kingdom=="Archaea"),]  # Subset from OTU_VST_resp_tax_cols  #Arch_resp
# Arch_corr
```

```r
# Improved heatmap data with COLORS for Taxa, from barplot
# Data preparation

# Get Taxonomy colors from above, note rename for merge  # OTU_VST_respM_agg2 # agg2taxCols
agg2taxCols<-data.frame(Phlya_cat = OTU_VST_respM_agg2$Tax2, cols) 

# Remake Phyla_cat # Copied from above, now column, FIX above?
OTU_VST_resp_tax$Phlya_cat <- ifelse(OTU_VST_resp_tax$Phylum == "Proteobacteria", paste(OTU_VST_resp_tax$Class), paste(OTU_VST_resp_tax$Phylum)) # Phlya_cats #OTU_VST_resp_tax

# Merge Responsive OTUs and colors by Phyla_cat
OTU_VST_resp_tax_cols <- merge(OTU_VST_resp_tax, agg2taxCols, by="Phlya_cat", all.x=TRUE)  #OTU_VST_resp_tax_cols

# Fix Tax2 names, replacing NAs 
OTU_VST_resp_tax_cols$Tax2 <- ifelse(is.na(OTU_VST_resp_tax_cols$cols), paste("Other"), paste(OTU_VST_resp_tax_cols$Phlya_cat)) # Phlya_cats # OTU_VST_resp_tax_cols

# Fix colors for "Other, replacing NAs
OtherCol=agg2taxCols$cols[agg2taxCols$Phlya_cat=="Other"] # Get color for "Other" # OtherCol
OTU_VST_resp_tax_cols$col2 <- ifelse(is.na(OTU_VST_resp_tax_cols$cols), paste(OtherCol), paste(OTU_VST_resp_tax_cols$cols)) # Phlya_cats # OTU_VST_resp_tax_cols # OTU_VST_resp_tax_cols

# Get color vector
TaxCol2<-OTU_VST_resp_tax_cols$col2     #TaxCol2

# Get data matrix (no row labels)
RespTaxM <-data.matrix(OTU_VST_resp_tax_cols[3:(ncol(OTU_VST_resp_tax_cols)-10)])  # RespTaxM   
                                 
# Get Z-scores, log2
RespTaxMZ <- data.matrix(t(scale(t(log2(RespTaxM)), center=TRUE, scale=TRUE)))     # z-scores, log10, by row 
```

```r
# Plot heatmap -- Too BIG
options(repr.plot.width=10, repr.plot.height=6)
heatmap.2(t(RespTaxMZ), Rowv=F, Colv=F, key=TRUE, trace="none", density.info="none",scale="column", ColSideColors = TaxCol2, margins = c(10, 10), col = rev(brewer.pal(11,"RdYlBu")))
# ColSideColors=trt_colors, ColSideColors = L2_colors,trace="none",
```
